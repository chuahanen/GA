````python
filter_options = {
    'Speed':   70,
    'Attack':  85,
    'Type':     'ElectricFlying',
    #'HP' : 100,
    'Defense' : 50,
    #'SpecialAttack': 70, 
    #'SpecialDefense': 50,
    #'Total' : 600
    }
````

````python
def pop_filter(filter_options):
    
    ''' 
    Function aims to start with a full list of all pokemons, and through each iteration, 
    filter away results that do not meet the specified filter criteria. ''' 
    
    results = list(pokedex.values())[:]
    criteria = list(filter_options.keys())
    
    for c in criteria:
        for row in pokedex:
            try:
                #for continuous variables, filter away less than or equals to criteria 
                if isinstance(filter_options[c],int) and float(pokedex[row][c]) <= filter_options[c]:
                    results.remove(pokedex[row])
                #for string, filter away non-exact matches
                elif isinstance(filter_options[c],str) and (pokedex[row][c]) != filter_options[c]:
                    results.remove(pokedex[row])
                #for everything else, keep
                else:
                    continue
            except:
                continue

    return results
````
